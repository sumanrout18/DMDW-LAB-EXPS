```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
df=pd.read_csv('Pokemon.csv',encoding='unicode_escape')
df.head()
```

```python
sb.lmplot(x='Attack', y='Defense', data=df,hue='Generation',fit_reg=True)
plt.show()
```

```python
sb.boxplot(df['Speed'],notch=True,orient='v',color='Green',width=0.74)
plt.show()
```

```python
plt.hist(df['Speed'],bins=20,ec='White',color='HotPink')
plt.show()
```

```python
sb.jointplot(data=df,x='Attack',y='Defense',kind='hex',color='Red')
plt.show()
```

```python
name=df['Name']
name=name[:10]
attack=df['Attack']
attack=attack[:10]
```

```python
plt.barh(name,attack,linewidth=0.7,color=['Red','Yellow','Green','Blue','Gold','Orange','Pink','Black'
,'Cyan','Brown'])
```

```python
generation=df['Generation'].unique()
plt.pie(generation,shadow=True,startangle=90,autopct='%1.2f%%')
plt.title('GENERATION WISE DATA')
plt.axis('equal')
plt.show()
```

```python
plt.plot(df['Attack'],color='Green')
plt.show()
```

```python
plt.scatter(df['Attack'],df['Sp. Atk'],color=['Gold'])
plt.xlabel('ATTACK OF POKEMON')
plt.ylabel('SPECIAL ATTACK OF POKEMON')
plt.title('ATTACK VS SPECIAL ATTACK')
plt.show()
```
